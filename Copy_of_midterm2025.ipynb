```python
interview = pd.read_csv('/content/drive/MyDrive/data midterm/interview.csv')
person = pd.read_csv('/content/drive/MyDrive/data midterm/person.csv')
print("ไฟล์ interview.csv และ person.csv โหลดสำเร็จแล้ว")
```

```python

# กรองหาคำให้การที่มีเบาะแสเกี่ยวกับคดีฆาตกรรม
# เราสามารถใช้คำสำคัญจากรายงานคดีหรือตรวจสอบเบาะแสที่เกี่ยวข้อง
witness_interview = interview[
    interview['transcript'].str.contains('gunshot') &
    interview['transcript'].str.contains('Get Fit Now Gym')
]
```

```python
if not witness_interview.empty:
    print("\nคำให้การของพยานที่เกี่ยวข้อง:")
    print(witness_interview)

    # ดึง person_id ของพยาน
    witness_ids = witness_interview['person_id'].tolist()
    print(f"\nรหัสประจำตัวของพยาน: {witness_ids}")

    # ค้นหาข้อมูลของพยานจากไฟล์ person.csv
    witness_info = person[person['id'].isin(witness_ids)]
    print("\nข้อมูลพยาน:")
    print(witness_info)
```

```python
get_fit_now_member = pd.read_csv('/content/drive/MyDrive/data midterm/get_fit_now_member.csv')
drivers_license = pd.read_csv('/content/drive/MyDrive/data midterm/drivers_license.csv')
print("ไฟล์ get_fit_now_member.csv และ drivers_license.csv โหลดสำเร็จแล้ว")
```

```python
# กรองหาสมาชิก Get Fit Now ที่มีรหัสขึ้นต้นด้วย '48Z' และเป็นสมาชิกแบบ 'gold'
suspect_gym = get_fit_now_member[
    (get_fit_now_member['id'].str.startswith('48Z')) &
    (get_fit_now_member['membership_status'] == 'gold')
]
```

```python
print("\nผู้ต้องสงสัยจากข้อมูลสมาชิกยิม:")
print(suspect_gym)
```

```python
suspect_car = drivers_license[
    drivers_license['plate_number'].str.contains('H42W')
]
```

```python
print("\nผู้ต้องสงสัยจากข้อมูลป้ายทะเบียนรถ:")
print(suspect_car)
```

```python
# รวมข้อมูลเพื่อระบุตัวตนผู้ต้องสงสัย
# เราจะรวมข้อมูลจาก person.csv กับข้อมูลใบขับขี่ของผู้ต้องสงสัย
person_from_car = pd.merge(suspect_car, person, left_on='id', right_on='license_id')
display(person_from_car)
```

```python
# จากนั้นรวมข้อมูลของคนที่มีทั้งป้ายทะเบียนรถและเป็นสมาชิกยิม
murderer_suspect = pd.merge(person_from_car, suspect_gym, left_on='id_y', right_on='person_id')

if not murderer_suspect.empty:
    print("\nข้อมูลผู้ต้องสงสัยฆาตกร:")
    print(murderer_suspect)
    murderer_name = murderer_suspect['name_y'].iloc[0]
    print(f"\nฆาตกรน่าจะเป็น: {murderer_name}")
```

#**หาตัวผู้บงการ**

```python
import pandas as pd

try:
    person = pd.read_csv('/content/drive/MyDrive/data midterm/person.csv')
    drivers_license = pd.read_csv('/content/drive/MyDrive/data midterm/drivers_license.csv')
    print("โหลดไฟล์ข้อมูลสำเร็จแล้ว")
except FileNotFoundError:
    print("ไม่พบไฟล์ที่จำเป็น กรุณาตรวจสอบว่าอัปโหลดไฟล์ครบถ้วนแล้ว")
```

```python
# ค้นหาบุคคลที่มีที่อยู่บนถนน Northwestern Dr
suspect_person = person[
    (person['address_street_name'] == 'Northwestern Dr')
]

if not suspect_person.empty:
    print("\nผู้ต้องสงสัยจากที่อยู่:")
    print(suspect_person)
```

```python
# รวมข้อมูลกับไฟล์ drivers_license เพื่อยืนยันลักษณะของผู้บงการ
mastermind_suspect = pd.merge(
    suspect_person,
    drivers_license,
    left_on='license_id',
    right_on='id'
)
display(mastermind_suspect)
```

```python
if not mastermind_suspect.empty:
    # กรองหาผู้หญิงผมสีแดง
    the_mastermind = mastermind_suspect[
        (mastermind_suspect['gender'] == 'female') &
        (mastermind_suspect['hair_color'] == 'red')
    ]

    if not the_mastermind.empty:
        mastermind_name = the_mastermind['name'].iloc[0] # Corrected from 'name_x' to 'name'
        print("\nข้อมูลผู้บงการ:")
        print(the_mastermind)
        print(f"\nผู้บงการในคดีนี้คือ: {mastermind_name}")
```
