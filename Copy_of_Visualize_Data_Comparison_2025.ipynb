```python
data1 = pd.read_csv('/content/drive/MyDrive/Python&Datavize2025/Data/2025-01/2025-01-bedfordshire-stop-and-search.csv')
data2 = pd.read_csv('/content/drive/MyDrive/Python&Datavize2025/Data/2025-02/2025-02-bedfordshire-stop-and-search.csv')
data3 = pd.read_csv('/content/drive/MyDrive/Python&Datavize2025/Data/2025-03/2025-03-bedfordshire-stop-and-search.csv')
data4 = pd.read_csv('/content/drive/MyDrive/Python&Datavize2025/Data/2025-04/2025-04-bedfordshire-stop-and-search.csv')
```

```python
compared_police_eth1 = data1.groupby('Officer-defined ethnicity')['Type'].count().reset_index()
compared_police_eth2 = data2.groupby('Officer-defined ethnicity')['Type'].count().reset_index()
compared_police_eth3 = data3.groupby('Officer-defined ethnicity')['Type'].count().reset_index()
compared_police_eth4 = data4.groupby('Officer-defined ethnicity')['Type'].count().reset_index()
```

```python
def swap_rows(tab, eth1='Asian', eth2='Other'):
    # Find the indices of 'Asian' and 'Other'
    asian_index = tab[tab['Officer-defined ethnicity'] == eth1].index[0]
    other_index = tab[tab['Officer-defined ethnicity'] == eth2].index[0]

    # Get the rows
    asian_row = tab.loc[asian_index].copy()
    other_row = tab.loc[other_index].copy()

    # Swap the rows
    tab.loc[asian_index] = other_row
    tab.loc[other_index] = asian_row

    return tab
```

```python
compared_police_eth1 = swap_rows(compared_police_eth1)
compared_police_eth2 = swap_rows(compared_police_eth2)
compared_police_eth3 = swap_rows(compared_police_eth3)
compared_police_eth4 = swap_rows(compared_police_eth4)
```

## Plot grouped bar chart ของ หนังอินเดีย โดย แยกกลุ่มตามประเภทหนัง และเปรียบเทียบ rating จาก ความยาวของหนัง

```python
data_india = pd.read_csv(os.path.join(path,'IMDb Movies India.csv'), encoding='latin-1')
data_india.head()
```

```python
data_india = data_india.dropna(subset=['Rating','Duration','Genre'])
data_india = data_india[['Rating','Duration','Genre']]
data_india
```

```python
# prompt: แปลงค่าใน column duration เป็น int โดยลบ min ออกและ เปลี่ยนชนิดข้อมูลเป็น integer

# Remove " min" from the 'Duration' column and convert to integer
data_india['Duration'] = data_india['Duration'].str.replace(' min', '').astype(int)
```

```python
data_india.describe()
```

```python
# prompt: แยกตารางออกเป็นตารางย่อย 5 ตารางสำหรับ genre: drama, action, romance, horror, musical

# Create a dictionary to store the tables
genre_tables = {}

# Define the genres to extract
genres = ['Drama', 'Action', 'Romance', 'Horror', 'Musical']

# Iterate through the genres and create separate tables
for genre in genres:
  genre_tables[genre] = data_india[data_india['Genre'] == genre].copy()

# Access the tables using the genre as the key
# Example:
# drama_table = genre_tables['Drama']
```

```python
all_table = pd.concat(genre_tables.values())
all_table
```

```python
# prompt: แบ่ง duration ออกเป็น short, medium, long โดย short สั้นกว่า 60 medium 60-90 ที่เหลือ เป็น long

# Function to categorize duration
def categorize_duration(duration):
  if duration < 60:
    return 'short'
  elif 60 <= duration <= 90:
    return 'medium'
  else:
    return 'long'

# Apply the function to create a new column
all_table['Duration Category'] = all_table['Duration'].apply(categorize_duration)
all_table
```

```python
all_table.groupby(['Genre','Duration Category'])['Rating'].mean()
```

```python
india_bar_data = all_table.groupby(['Genre','Duration Category'])['Rating'].mean().unstack()
```

```python
india_bar_data
```

```python
ax = india_bar_data.plot(kind='bar', figsize=(8, 6), width=0.8)

# Add labels and title
plt.xlabel('Genre')
plt.ylabel('Rating')
plt.title('หนังอินเดีย โดย แยกกลุ่มตามประเภทหนัง และเปรียบเทียบ rating จาก ความยาวของหนัง')

# Move legend outside the plot
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Adjust plot size
plt.gcf().set_size_inches(plt.gcf().get_size_inches() * 0.7)

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd  #import pandas มาเพราะข้อมูลจะเกี่ยวกับตาราง
import matplotlib.pyplot as plt
from math import pi #import math มาเนื่องจากเกี่ยวกับวงกลมนำมาใช้เพื่อดึงค่าพาย
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data for two cities using the correct path
df_bedfordshire = pd.read_csv("/content/drive/MyDrive/Python&Datavize2025/Data/2025-01/2025-01-bedfordshire-stop-and-search.csv")
df_cambridge = pd.read_csv("/content/drive/MyDrive/Python&Datavize2025/Data/2025-01/2025-01-cambridgeshire-stop-and-search.csv")

# Count crime types and normalize
crime_bedfordshire = df_bedfordshire['Object of search'].value_counts(normalize=True) * 100
crime_cambridge = df_cambridge['Object of search'].value_counts(normalize=True) * 100

# Combine crime types from both cities and select top N (handling potential NaNs)
crime_total = (crime_bedfordshire.add(crime_cambridge, fill_value=0)).sort_values(ascending=False)

# Filter out crime types with no occurrences in either city after combining
crime_total = crime_total[crime_total > 0]

# Select top 5 crime types (or fewer if less than 5 unique crime types exist)
top_n = 5
top_crimes = crime_total.head(top_n).index.tolist()


# Prepare data for radar chart (%) - handling potential missing crime types in individual cities
data_bedfordshire = [crime_bedfordshire.get(c, 0) for c in top_crimes]
data_cambridge = [crime_cambridge.get(c, 0) for c in top_crimes]


# Close the circle for the radar chart
data_bedfordshire += data_bedfordshire[:1]
data_cambridge += data_cambridge[:1]


# Angles for plotting
angles = np.linspace(0, 2*np.pi, len(top_crimes), endpoint=False).tolist()
angles += angles[:1]

# --- Radar Chart ---
fig, ax = plt.subplots(figsize=(8,8), subplot_kw=dict(polar=True))
ax.plot(angles, data_bedfordshire, linewidth=2, label="Bedfordshire")
ax.fill(angles, data_bedfordshire, alpha=0.25)

ax.plot(angles, data_cambridge, linewidth=2, label="Cambridgeshire")
ax.fill(angles, data_cambridge, alpha=0.25)


ax.set_xticks(angles[:-1])
ax.set_xticklabels(top_crimes, fontsize=10)
ax.set_title("Top {} Crime Types (% Comparison)".format(top_n), fontsize=14)
ax.legend(loc="upper right", bbox_to_anchor=(1.3,1.1))
plt.show()

# --- Summary Table (%) ---
summary_df = pd.DataFrame({
    "Crime Type": top_crimes,
    "Bedfordshire (%)": [crime_bedfordshire.get(c, 0) for c in top_crimes],
    "Cambridgeshire (%)": [crime_cambridge.get(c, 0) for c in top_crimes]
})

print("\nTop {} Crime Types (% Comparison)".format(top_n))
print(summary_df.round(2))
```
